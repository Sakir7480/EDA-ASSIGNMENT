9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot.

.To visually explore how track duration varies across different artists, we can use a box plot or a violin plot. These plots help compare distributions and identify variations in track duration among different artists.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("your_file.csv")  # Replace with your actual file

# Check if required columns exist
if 'artist' in df.columns and 'duration_ms' in df.columns:
    # Convert duration from milliseconds to minutes for better readability
    df['duration_min'] = df['duration_ms'] / (1000 * 60)

    # Select the top 10 artists with the most tracks for better visualization
    top_10_artists = df['artist'].value_counts().nlargest(10).index
    df_filtered = df[df['artist'].isin(top_10_artists)]

    # Set plot size
    plt.figure(figsize=(12, 6))

    # Box Plot
    sns.boxplot(x='artist', y='duration_min', data=df_filtered, palette="coolwarm")
    plt.xticks(rotation=45)  # Rotate artist names for readability
    plt.xlabel("Artist")
    plt.ylabel("Track Duration (Minutes)")
    plt.title("Variation of Track Duration Across Top 10 Artists")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

    # Violin Plot (optional)
    plt.figure(figsize=(12, 6))
    sns.violinplot(x='artist', y='duration_min', data=df_filtered, palette="coolwarm", inner="quartile")
    plt.xticks(rotation=45)
    plt.xlabel("Artist")
    plt.ylabel("Track Duration (Minutes)")
    plt.title("Distribution of Track Duration Across Top 10 Artists")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

else:
    print("Columns 'artist' or 'duration_ms' not found in the dataset.")
```

10..How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.

.To visualize how the distribution of track popularity varies among different artists, we can use a swarm plot or a violin plot. These plots help compare popularity distributions across multiple artists.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("your_file.csv")  # Replace with your actual file

# Check if required columns exist
if 'artist' in df.columns and 'popularity' in df.columns:
    # Select the top 10 artists with the most tracks for better visualization
    top_10_artists = df['artist'].value_counts().nlargest(10).index
    df_filtered = df[df['artist'].isin(top_10_artists)]

    # Set plot size
    plt.figure(figsize=(12, 6))

    # Swarm Plot
    sns.swarmplot(x='artist', y='popularity', data=df_filtered, palette="viridis", size=5, dodge=True)
    plt.xticks(rotation=45)  # Rotate artist names for readability
    plt.xlabel("Artist")
    plt.ylabel("Track Popularity")
    plt.title("Distribution of Track Popularity Across Top 10 Artists")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

    # Violin Plot (optional)
    plt.figure(figsize=(12, 6))
    sns.violinplot(x='artist', y='popularity', data=df_filtered, palette="viridis", inner="quartile")
    plt.xticks(rotation=45)
    plt.xlabel("Artist")
    plt.ylabel("Track Popularity")
    plt.title("Distribution of Track Popularity Across Top 10 Artists")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

else:
    print("Columns 'artist' or 'popularity' not found in the dataset.")
```
